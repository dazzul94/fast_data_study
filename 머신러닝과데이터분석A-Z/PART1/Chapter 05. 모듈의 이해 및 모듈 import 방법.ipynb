#### from import 
 - 해당 모듈에서 특정한 타입만 import

```python
1 + 2
3 - 1
3 * 5
```

```python
import math

math.pi

math.cos(100)
```

```python
from math import pi # math에서 특정한 pi만 import 하겠다.

pi
```

```python
from math import cos # math에서 특정한 cos만 import 하겠다.

cos(100)
```

#### \* 임포트
 - 해당 모듈내에 정의된 모든 것을 import
 - 일반적으로 사용이 권장되지 않음
     + 왜? 이름에 중복이 있을 수 있기 때문이다.

```python
from math import * # math 안의 모든 기능을 임포트 한다.
```

```python
sin(100)

e
```

#### as 
 - 모듈 import 시, alias(별명) 지정가능
